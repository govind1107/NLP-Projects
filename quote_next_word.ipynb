```python
total_words = len(tokenizer.word_index) + 1
```

```python
input_sentences = []
for sentences in text.split("."):
  tokenized_sentences = tokenizer.texts_to_sequences([sentences])[0]

  for i in range(len(tokenized_sentences)):
    input_sentences.append(tokenized_sentences[:i+1])

```

```python

max_len = max([len(x) for x in input_sentences])
from tensorflow.keras.preprocessing.sequence import pad_sequences
padded_input_sequences = pad_sequences(input_sentences, maxlen = max_len, padding='pre')
```

```python

X = padded_input_sequences[:,:-1]
y = padded_input_sequences[:,-1]
```

```python
y = np.array(tf.keras.utils.to_categorical(y, num_classes=total_words))
```

```python
model = Sequential()
model.add(Embedding(total_words, 100, input_length=max_len-1))
model.add(LSTM(150))
model.add(Dense(total_words, activation='softmax'))
print(model.summary())
```
